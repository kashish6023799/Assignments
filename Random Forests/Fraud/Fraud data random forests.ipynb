```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import preprocessing
from sklearn import metrics
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
fraud=pd.read_csv("Fraud_check.csv")
```

```python
fraud
```

```python
fraud=fraud.rename(columns={"Marital.Status":"Marital","Taxable.Income":"Tax_income","City.Population":"City_population","Work.Experience":"Work_exp"})
fraud
```

```python
fraud["Income_Status"]=np.where(fraud["Tax_income"]>=30000,"Good","Risky")# 1 is Good and 0 is Risky
fraud.head(10)
```

```python
label_encoder=preprocessing.LabelEncoder()
fraud['Undergrad']=label_encoder.fit_transform(fraud['Undergrad'])
fraud['Marital']=label_encoder.fit_transform(fraud['Marital'])
fraud['Urban']=label_encoder.fit_transform(fraud['Urban'])
fraud['Income_Status']=label_encoder.fit_transform(fraud['Income_Status'])
```

```python
fraud.drop(["Tax_income"],axis=1,inplace=True)
```

```python
fraud.copy()
```

```python
x=fraud.iloc[:,:5]
y=fraud["Income_Status"]
```

```python
x
```

```python
y
```

```python
# Splitting the data into training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Using random forests classification

```python

num_trees=100
max_features=6
KFold=KFold(n_splits=10)

model=RandomForestClassifier(n_estimators=num_trees,max_samples=0.7,max_features=max_features,random_state=7)
results=cross_val_score(model,x,y,cv=KFold)
print(results.mean())
```

# Build random forests classification using gini

```python
from sklearn.model_selection import KFold
num_trees=100
max_features=6
KFold=KFold(n_splits=10)

model_gini=RandomForestClassifier(n_estimators=num_trees,max_features=max_features,criterion="gini",random_state=7)
results_gini=cross_val_score(model_gini,x,y,cv=KFold)
print(results_gini.mean())
```

# Build random forests classification using Entropy

```python

from sklearn.model_selection import KFold
num_trees=100
max_features=6
KFold=KFold(n_splits=12)

model_en=RandomForestClassifier(n_estimators=num_trees,max_samples=0.8,max_features=max_features,criterion="entropy",min_samples_split=5,random_state=10)
results_en=cross_val_score(model_en,x,y,cv=KFold)
print(results_en.mean())
```
