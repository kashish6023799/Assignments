```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib as plt
import string
import spacy
import re
from matplotlib.pyplot import imread
%matplotlib inline
```

```python
tweets=pd.read_csv("Elon_musk.csv",encoding="Latin-1")
tweets=tweets.drop(["Unnamed: 0"],axis=1)
tweets.head()
```

```python
tweets=[Text.strip() for Text in tweets.Text] # remove both the leading and the trailing characters
tweets=[Text for Text in tweets if Text] # removes empty strings, because they are considered in pthon as false
tweets[0:10]
```

```python
#joining the list into one string/text
tweets_text=' '.join(tweets)
tweets_text
```

```python
#remove twitter handler name(Remove @username)
from nltk.tokenize import TweetTokenizer
tkn= TweetTokenizer(strip_handles=True)
tweets_tokens=tkn.tokenize(tweets_text)
print(tweets_tokens)
```

```python
#agian joining the list into one string/text
tweets_tokens_text=" ".join(tweets_tokens)
tweets_tokens_text
```

```python
#remove punctuations
no_punc=tweets_tokens_text.translate(str.maketrans('','',string.punctuation))
no_punc
```

```python
#remove url and http text
no_url=re.sub(r'https\S+','',no_punc)
no_url
```

# Named Entity Recognition (NER)

```python
#Part of Speech (pos) Tagging, NER is Used for recognise the word what typr of word it is 
nlp=spacy.load('en_core_web_sm')

one_block=clean_tweets
doc_block=nlp(one_block)
spacy.displacy.render(doc_block,style='ent',jupyter=True)
```

```python
for token in doc_block[100:200]:
    print(token,token.pos_)
```

```python
#Filtering the nouns and verbs only
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[100:200])
```

```python
# Counting noun and verb token using countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq,key=lambda x:x[1],reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10] #viewing top ten results
```

```python
#Visualizing results (barchart for top 10 nouns and verbs)
wd_df[0:10].plot.bar(x='word',figsize=(12,8),title="top 10 nouns and verbs");
```

# Implementing the Sentiment Analysis

```python
from nltk import tokenize
sentences=tokenize.sent_tokenize(' '.join(tweets))
sentences
```

```python
sent_df=pd.DataFrame(sentences,columns=['sentence'])
sent_df
```

```python
#Emotion Lexicon-Affin
affin=pd.read_csv("Afinn.csv",sep=',',encoding='Latin-1')
affin
```

```python
affinity_scores=affin.set_index('word')['value'].to_dict()
affinity_scores
```

```python
#creating function: score each word in a sentence in lemmatised form, but calculate the score for the whole original sentences
nlp=spacy.load('en_core_web_sm')
sentiment_lexicon=affinity_scores

def calculate_sentiment(text:str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score+=sentiment_lexicon.get(word.lemma_,0)
        return sent_score
```

```python
#manual testing to check positive and negative words
calculate_sentiment(text='bad')
```

```python
#Calculating sentiment value for each sentence
sent_df['sentiment_value']=sent_df['sentence'].apply(calculate_sentiment)
sent_df['sentiment_value']
```

```python
#how many words are there in a sentence?
sent_df['word_count']=sent_df['sentence'].str.split().apply(len)
sent_df['word_count']
```

```python
sent_df.sort_values(by='sentiment_value')
```

```python
#Sentiment score of the whole review
sent_df['sentiment_value'].describe()
```

```python
#negative sentiment score of the whole review
sent_df[sent_df['sentiment_value']<=0]
```

```python
#positive sentiment score of the whole review
sent_df[sent_df['sentiment_value']>0]
```
